```
def isPrime(a):
  if(a<2):
    return False
  for i in range(2,a):
    if(a%i==0):
      return False
  return True

A = input().split()
B = input().split()
result = 0
for i in range(len(A)):
  prime=[]
  pdA=[]
  N = int(A[i])
  for j in range(1,N+1):
    if(isPrime(j) == True):
      prime.append(j)
  for j in range(len(prime)):
    if(int(A[i])%prime[j] == 0):
      pdA.append(prime[j])
  pdA = set(pdA)
  prime=[]
  pdB=[]
  N = int(B[i])
  for j in range(1,N+1):
    if(isPrime(j) == True):
      prime.append(j)
  for j in range(len(prime)):
    if(int(B[i])%prime[j] == 0):
      pdB.append(prime[j])
  pdB = set(pdB)
  if(pdA == pdB):
    result += 1
print(result)
```
